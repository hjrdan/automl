```python
# Resource monitoring setup:
process = psutil.Process()  # Get current process
start_time = time.time()

#monitor
start_memory = process.memory_info().rss  # Start memory usage
start_cpu_percent = psutil.cpu_percent(interval=None)  # Start CPU usage
start_cpu_time = process.cpu_times()  # Start CPU time

# Start:
# Configuration 1: All model types included
aml1 = H2OAutoML(exclude_algos=None, max_models = 50, seed=42, verbosity="info")
aml1.train(x=x, y=y, training_frame = train, validation_frame=valid)

# End resource monitoring:
end_time = time.time()
end_memory = process.memory_info().rss
end_cpu_percent = psutil.cpu_percent(interval=None)
end_cpu_time = process.cpu_times()

# Calculate resource usage:
elasped_time = end_time - start_time
memory_usage = (end_memory - start_memory) /(1024**2)
avg_cpu_percent = (end_cpu_percent + start_cpu_percent) / 2
cpu_time_usage = end_cpu_time.system - start_cpu_time.system


print(f"H2O execution time: {elasped_time:.2f} seconds")
print(f"H2O memory usage: {memory_usage:.2f} MB")
print(f"H2O Average CPU usage: {avg_cpu_percent:.2f}%")
print(f"H2O CPU time usage: {cpu_time_usage:.2f} seconds")

# View leaderboard
lb = aml1.leaderboard
print(lb)
```

```python
print(f"H2O execution time: {elasped_time:.2f} seconds")
memory_usage = (end_memory) /(1024**2)
print(f"H2O memory usage: {memory_usage:.2f} MB")
print(f"H2O Average CPU usage: {avg_cpu_percent:.2f}%")
print(f"H2O CPU time usage: {cpu_time_usage:.2f} seconds")
```

```python
# Get the cross-validation leaderboard from the H2OAutoML object
lb3cv = aml_cv3.leaderboard

# Print the cross-validation results
print(lb3cv)

# Inspect the individual fold results
for i in range(5):
    # Get the model ID for the current fold from the leaderboard
    model_id = lb3cv[i, 'model_id']
    # Use the model ID to get the actual model object using h2o.get_model()
    fold_model = h2o.get_model(model_id)
    # Get the model performance object for the current fold
    fold_performance = fold_model.model_performance()
    # Access specific metrics using their respective methods
    fold_metrics = {
        "accuracy": fold_performance.accuracy()[0] if fold_performance.accuracy() else None, # Handle case where accuracy is not available
        "precision": fold_performance.precision()[0] if fold_performance.precision() else None, # Handle case where precision is not available
        "recall": fold_performance.recall()[0] if fold_performance.recall() else None, # Handle case where recall is not available
        "f1": fold_performance.F1()[0] if fold_performance.F1() else None # Handle case where F1 is not available
    }
    print(f"Fold {i+1} Metrics:")
    print(f"Accuracy: {fold_metrics['accuracy']}")
    print(f"Precision: {fold_metrics['precision']}")
    print(f"Recall: {fold_metrics['recall']}")
    print(f"F1-Score: {fold_metrics['f1']}")
    print()
```
